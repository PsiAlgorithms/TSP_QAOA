```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
with open("data/coordinates.pkl", "rb") as fp:
    coordinates = pkl.load(fp)
```

```python
fig, ax = plt.subplots(5, 2, figsize=(8, 15))
fig.tight_layout()
ax = ax.flatten()
for cr in range(len(coordinates)):
    x = [x for x, _ in coordinates[cr]]
    y = [x for _, x in coordinates[cr]]
    ax[cr].scatter(x[0], y[0], marker="s", color="red")
    ax[cr].scatter(x[1:], y[1:], color="royalblue")
    ax[cr].grid(True, linestyle="dotted", alpha=0.5)
    # ax[cr].set_xlim([0,100])
    # ax[cr].set_ylim([0,100])
    ax[cr].set_title(str(cr))
```
